```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

folder_path = 'datasets/'
df = pd.read_csv('datasets/athlete_events.csv')
regions = pd.read_csv(folder_path+'noc_regions.csv')
df = df.merge(regions, left_on='NOC', right_on='NOC')
df.head()
```

```python
gdp = pd.read_csv(folder_path+'gdp.csv')
gdp = gdp[['Entity','Year','GDP per capita']]

health_exp = pd.read_csv(folder_path+'healthcare_expenditure_gdp.csv')
health_exp.drop(columns={'Code'}, inplace=True)

obesity = pd.read_csv(folder_path+'obesity_adults.csv')
obesity.drop(columns={'Code'}, inplace=True)

regime = pd.read_csv(folder_path+'political_regime.csv')
regime.drop(columns={'Code'}, inplace=True)
mapping = {0: 'closed autocracy', 1: 'electoral autocracy', 2: 'electoral democracy', 3: 'liberal democracy'}
regime['Political regime'] = regime['Political regime'].replace(mapping)

poverty = pd.read_csv(folder_path+'poverty.csv')
poverty.rename(columns={'Share below $1 a day':'1$', 'Share below $2.15 a day':'2.15$', 'Share below $3.65 a day':'3.65$', 'Share below $10 a day':'10$'}, inplace= True)
poverty = poverty[['Country', 'Year', '1$', '2.15$', '3.65$', '10$']]
```

```python
population = pd.read_csv(folder_path+'population.csv')
population.drop(columns={'Code'}, inplace=True)
population.rename(columns={'Population (historical estimates)':'Population'}, inplace=True)
population['20_y_bef'] = population.groupby('Entity')['Population'].transform(lambda x: x.shift(3))

pop_500_y_bef = population.loc[(population['Year']>1590)&(population['Year']<1650)]
pop_500_y_bef = pop_500_y_bef.groupby('Entity')['Population'].max().reset_index()

pop_100_y_bef = population.loc[(population['Year']>1950)&(population['Year']<1960)]
pop_100_y_bef = pop_100_y_bef.groupby('Entity')['Population'].max().reset_index()

pop = population.sort_values(by='Year',ascending=False)
pop.drop_duplicates(subset='Entity', keep='first')
pop.reset_index(drop=True, inplace=True)

population = population.loc[population['Year']>1890]
```

```python
df = df[(~df['Medal'].isna())&(df['Season']=='Summer')]
df = df.groupby(['Year','region'])['Medal'].count().reset_index().sort_values(by=['Year','region'], ascending=False)
df['Tot_med'] = df.groupby(['region'])['Medal'].cumsum()
df = df[df['Medal']>10]
sns.displot(df['Medal'], kde= True)
plt.show()
```

```python
merged_df = gdp.merge(health_exp, how='outer', on=['Entity','Year'])
merged_df = merged_df.merge(obesity, how='outer', on=['Entity','Year'])
merged_df = merged_df.merge(regime, how='outer', on=['Entity','Year'])
merged_df = merged_df.merge(population, how='outer', on=['Entity','Year'])
merged_df = merged_df.merge(poverty, how='outer', left_on=['Entity','Year'], right_on=['Country','Year'])
merged_df = merged_df.merge(df, how='outer', left_on=['Entity','Year'], right_on=['region','Year'])

merged_df.head()
merged_df = merged_df.loc[merged_df['Year']>1890]
merged_df.drop(columns={'Country', 'region'}, inplace=True)
df_1_0 = merged_df.loc[(merged_df['Year']>1965)&(merged_df['Year']<2017)]
df_1_0.rename(columns={'GDP per capita':'GDP', 
                       'Current health expenditure (CHE) as percentage of gross domestic product (GDP) (%)':'Health exp', 
                       'Prevalence of obesity among adults, BMI >= 30 (crude estimate) (%) - Sex: both sexes - Age group: 18+  years':'Obesity',
                       '20_y_bef':'Population_20_y_bef'}, inplace=True)

df_1_0.head()
```

```python
print(df_1_0['Medal'].isna().size)
print(df_1_0.size)
df_1_0 = df_1_0.loc[~df_1_0['Medal'].isna()]
df_1_0.size
sns.pairplot(df_1_0[['Entity','1$', '2.15$', '3.65$', '10$', 'Population', 'Population_20_y_bef', 'GDP', 'Health exp', 'Obesity', 'Political regime', 'Medal','Tot_med']])

plt.show()
```

```python
import numpy as np

correlation_matrix = df_1_0[['1$', '2.15$', '3.65$', '10$', 'Population', 'Population_20_y_bef', 'GDP', 'Health exp', 'Obesity', 'Medal','Tot_med']].corr()
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
plt.figure(figsize = (15,11))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
